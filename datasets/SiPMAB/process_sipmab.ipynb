```python
import pandas as pd
```

```python
sipmab_test = pd.read_csv('./sipmab_test_sample.csv')
sipmab_test
```

```python
# Добавление колонки chain между колонками mutation и ddG (на самом деле не важен порядок колонок)
sipmab_test['chain'] = sipmab_test['mutation'].str[0]
ddG_index = sipmab_test.columns.get_loc('ddG')
sipmab_test.insert(ddG_index, 'chain', sipmab_test.pop('chain'))
sipmab_test['mutation'] = sipmab_test['mutation'].str[1:]

# Добавление колонки class
sipmab_test['class'] = sipmab_test['ddG'].apply(lambda x: 1 if x > 0.21 else 0)

sipmab_test
```

```python
sipmab_test.to_csv('sipmab_test_sample_mod.csv', index=False)
```
